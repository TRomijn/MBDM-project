```python
import ema_workbench as ema
import pandas as pd
```

```python
all_policies = ema.load_results("Original Files/data/no policy 5000 runs.tar.gz")
```

```python
experiments, outcomes = all_policies
```

```python
outcomes['Costs'].shape
```

```python
new_dict = dict()
for key in outcomes.keys():
    values = list(outcomes[key][:,0])
    new_dict[key] = values
```

```python
df_outcomes = pd.DataFrame.from_dict(new_dict)
```

```python
df_outcomes
```

```python
df_experiments = pd.DataFrame.from_dict(experiments)
```

```python
no_policy_total = pd.concat([df_experiments, df_outcomes],axis=1)
```
