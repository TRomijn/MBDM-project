```python
def fix_dict(outcomes):
    new_dict = dict()
    for key in outcomes.keys():
        values = list(outcomes[key][:,0])
        new_dict[key] = values
    df_outcome = pd.DataFrame.from_dict(new_dict)
    return df_outcome
```
